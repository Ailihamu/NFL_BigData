I'll try to draw a football field first

```python
def draw_field(scrimmage=False):
    
    """
    Input:
    
    scrimmage - x axis of the line of scrimmage
    
    Output:
    fig
    ax 
    """
    
    
    fig=plt.figure()
    fig.set_size_inches(12, 5.33) #the figure size according to the ratio of a football field
    ax = fig.add_subplot()
    plt.ylim(0, 53.3) #height of the field
    plt.xlim(0, 120) #Width of the field
    
    #Draw a football field
    field = patches.Rectangle([0,0], width = 120, height = 53.3, color='green', linewidth=0.1, zorder=0,  alpha=0.5)
    
    #Draw two end zones
    RightEndZone = patches.Rectangle([0,0], width = 10, height = 53.3, color='darkblue', linewidth=0.1, zorder=0,  alpha=0.5)
    LeftEndZone = patches.Rectangle([110,0], width = 10, height = 53.3, color='darkblue', linewidth=0.1, zorder=0,  alpha=0.5)
    
    s = [field, RightEndZone, LeftEndZone]
    for i in s:
        ax.add_patch(i) #Add the patches to ax
        
    for x in ([10, 20, 30, 40, 50, 60]):
        ax.plot([x, x], [0, 53.3], color='white')
        ax.plot([50+x, 50+x], [0, 53.3], color='white') # Yard line
        
    for x in range(10,110):
        ax.plot([x, x], [0.4, 0.7], color='white')
        ax.plot([x, x], [53.0, 52.5], color='white')
        ax.plot([x, x], [22.91, 23.57], color='white')
        ax.plot([x, x], [29.73, 30.39], color='white') 
        
    for x in ([10, 20, 30, 40, 50, 60]):
        ax.plot([x-0.5, x+0.5], [23.34, 23.34], color='white')
        ax.plot([x-0.5, x+0.5], [30.06, 30.06], color='white') 
    
    for x in (20, 30, 40, 50, 60):
        plt.text(x-2, 6, str(x-10), color='white', fontsize='16')
        plt.text(x-2, 47.3, str(x-10), color='white', fontsize='16', rotation='180')
        plt.text(118-x, 6, str(x-10), color='white',fontsize='16')
        plt.text(118-x, 47.3, str(x-10), color='white', fontsize='16',rotation='180')
        
    if scrimmage:
        x = scrimmage
        plt.plot([x, x], [0, 53.3], color='red')
        plt.text(x+1, 50, '<-- the line of scrimmage', color='white') #line of scrimmage 
    
    return fig, ax

draw_field()
    
```

I'll try to plot each paly with it's unique 'Play' ID

```python
def plot_play(play_id, df=df):
    
    """
    Input:
    
    play_id - the unique PlayId of the specific play that we want to show
    df - the dataset 

    Output:
    show the football field and the players of two teams
    """
    
    play = df[df.PlayId==play_id] # choose a specific play to show
    
    if play.X[0] < 60:      
        x = int(play.YardLine[0]) - 10 # is the position of the Yardline on the left half of the field
    else:
        x = 110 - int(play.YardLine[0]) # x is the position of the Yardline on the right half of the field
    
    fig, ax = draw_field(scrimmage=x)
    
    x_home = play[play.Team=='home']['X']
    y_home = play[play.Team=='home']['Y']
    ax.scatter(x_home, y_home, color='Yellow', label='HomeTeam') #The home-team players
    
    x_away = play[play.Team=='away']['X']
    y_away = play[play.Team=='away']['Y']
    ax.scatter(x_away, y_away, color='orange', label='AwayTeam') #The away-team players
    
    x_away = play[play.NflId==play.NflIdRusher]['X']
    y_away = play[play.NflId==play.NflIdRusher]['Y']
    ax.scatter(x_away, y_away, color='black', label='Rusher') #The rusher 
    
    #the title of the figure
    ax.set_title(f'Play:{play_id} / Home team: {play.HomeTeamAbbr[0]} / Away Team: {play.VisitorTeamAbbr[0]} / Possesion = {play.PossessionTeam[0]} / Yard gained: {play.Yards[0]}')

    plt.legend()
    plt.show()
```
